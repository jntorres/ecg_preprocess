### Example of preprocessing of ecgs 

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
from preprocess_ecgs_00 import loadecg, notch, baseline_wander_removal
```

```python
path = "/home/users/jntorres/jessica/stanford_ECG_projects/ecgnet_hha/hha-ecgs-dir/Signals/"
file = 'e697214e72ad23cbe5ce.npy'


ecg = loadecg(path + file)
print(ecg.shape)
ecg_notch_removed = notch(ecg, 500)
ecg_baseline_removed = baseline_wander_removal(ecg, 500)
ecg_notch_baseline_removed = baseline_wander_removal(ecg_notch_removed, 500)
```

#### Following plots have the before depicted in the left column and the after on the right

#### Plotting only notch removal

```python

lead_order = ['1', '2', '3', 'R', 'L', 'F', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']

# Tracing plots 
columns = 2
rows = 12
fig, ax_array = plt.subplots(rows, columns,squeeze=False, figsize=(15,20))
for i,ax_row in enumerate(ax_array):
    for j,axes in enumerate(ax_row):
        if j == 0:
            signal1 = ecg[i,:] 
            x1 = np.linspace(0, 10, signal1.shape[0])
            y1 = np.squeeze(signal1)
            axes.plot(x1, y1, 'k-')
            axes.set_xlim(x1.min(), x1.max())
            axes.set_ylim(y1.min(), y1.max())
            axes.set_title('Lead ' + lead_order[i], size=20)
            axes.spines['top'].set_visible(False)
            axes.spines['right'].set_visible(False)
        else:
            signal2 = ecg_notch_removed[i,:]
            x1 = np.linspace(0, 10, signal2.shape[0])
            y1 = np.squeeze(signal2)
            axes.plot(x1, y1, 'k-')
            axes.set_title('Lead ' + lead_order[i], size=20)
            axes.set_xlim(x1.min(), x1.max())
            axes.set_ylim(y1.min(), y1.max())
            axes.spines['top'].set_visible(False)
            axes.spines['right'].set_visible(False)

plt.subplots_adjust(hspace=0.9)
plt.show()
```
